```python
import cv2
import google.colab.patches as colab
import numpy as np
import matplotlib.pyplot as plt
```

```python
inImage =cv2.imread("/content/drive/MyDrive/license_plate.bmp",0)
```

```python
grad_X =cv2.Sobel(inImage,cv2.CV_16S,1,0)
grad_y =cv2.Sobel(inImage,cv2.CV_16S,0,1)
```

```python
abs_grad_x =cv2.convertScaleAbs(grad_X)
abs_grad_y =cv2.convertScaleAbs(grad_y)
```

```python
print(abs_grad_x.shape)
print(abs_grad_y.shape)

grad_image =cv2.addWeighted(abs_grad_x ,1.0,abs_grad_y,1.0,0)

thrv, binImage=cv2.threshold(grad_image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

colab.cv2_imshow(binImage)
```
